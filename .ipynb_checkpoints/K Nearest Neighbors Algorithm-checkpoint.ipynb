```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
irish = load_iris()
irish.feature_names
```

```python
irish.target_names
```

```python
df = pd.DataFrame(irish.data, columns = irish.feature_names)
df
```

```python
df['target'] = irish.target
df
```

```python
print(df[df.target == 1].shape)
df[df.target == 1]
```

```python
print(df[df.target == 0].shape)
df[df.target == 0]
```

```python
print(df[df.target == 2].shape)
df[df.target == 2]
```

```python
df['flower_name'] = df.target.apply(lambda x : irish.target_names[x])
df
```

```python
df0= df[:50]
df1 = df[50 : 100]
df2 = df[100 : 150]
```

```python
sns.scatterplot(df0['sepal length (cm)'], df0['sepal width (cm)'], color='red', marker='+')
sns.scatterplot(df1['sepal length (cm)'], df1['sepal width (cm)'], color='green', marker='*')
```

```python
sns.scatterplot(df0['sepal length (cm)'], df0['sepal width (cm)'], color='red', marker='+')
sns.scatterplot(df1['sepal length (cm)'], df1['sepal width (cm)'], color='green', marker='*')
sns.scatterplot(df2['sepal length (cm)'], df2['sepal width (cm)'], color='blue', marker='.')
plt.title('Sepal length vs Sepal width (setosa, versicolor, virginicaf)')
plt.legend()
```

```python
sns.scatterplot(df0['petal length (cm)'], df0['petal width (cm)'], color='red', marker='+')
sns.scatterplot(df1['petal length (cm)'], df1['petal width (cm)'], color='green', marker='*')
sns.scatterplot(df2['petal length (cm)'], df2['petal width (cm)'], color='blue', marker='.')
plt.title('petal length vs petal width (setosa, versicolor, virginicaf)')
plt.legend()
```

```python
x = df.drop(['target', 'flower_name'], axis='columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain, xtest, ytrain, ytest = tts(x, y, test_size=0.25, random_state=1)
```

```python
print(len(xtrain), len(ytrain))
```

```python
print(len(xtest), len(ytest))
```

```python
from sklearn.neighbors import KNeighborsClassifier as kc
```

```python
knn = kc(n_neighbors=3)
knn.fit(xtrain, ytrain)
```

```python
knn.score(xtest, ytest)
```

```python
knn.score(xtrain, ytrain)
```

```python
df
```

```python
knn.predict([[7, 1.3, 5.9, 4]])
```

```python
ytest
```

```python
knn.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = knn.predict(xtest)

cm = confusion_matrix(ytest, y_pred)
cm
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, y_pred))
```
